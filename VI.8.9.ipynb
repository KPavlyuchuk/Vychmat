Мы знаем что на равномерной сетке ошибка оценивается сверху как:

$$E \leq \frac{M_{4}}{4!} \cdot \underset{[-4, -1]}{max}|(x+4)(x+3)(x+2)(x+1)|$$

```python
import matplotlib.pyplot as plt
import decimal
import numpy as np
%matplotlib inline
plt.rc('font', **{'size' : 22})

xlist = np.around(np.arange(-4, -1, 0.001), decimals=4)
ylist_my = (xlist + 4)*(xlist + 3)*(xlist + 2)*(xlist + 1)
plt.figure(figsize=(10, 5))


plt.plot(xlist, ylist_my)

plt.xlabel('Ось $x$')
plt.ylabel('Ось $y$')

plt.show()
```
